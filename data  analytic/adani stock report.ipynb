```python
#we will analyze  the year by year from 2008 
adani_2k8=ad_1.iloc[23:269]#short listing using the iloc
adani_2k8=adani_2k8.set_index(pd.DatetimeIndex(adani_2k8['Date'].values))
adani_2k8.drop(['Date'],inplace= True ,axis=1)
adani_2k8
#adani_2k8=adani.loc[adani[‘date’]]
#mask = (adani['2008-01-01'] > start_date) & (adani['2018-12-31'] <= end_date)
```

```python
#The pct_change() function in Pandas is used to calculate the percentage change
#between consecutive elements in a Pandas DataFrame or Series
returns_day2k8= adani_2k8['High'].pct_change()
returns_day2k8 
```

```python
#day returnes of 2k8
returns_daymax2k8 =returns_day2k8.max()
print(returns_daymax2k8,'max day volume')
returns_daymin2k8 =returns_day2k8.min()
print(returns_daymin2k8,'max day volume')
returns_dayavg2k8 =returns_day2k8.mean()
print(returns_dayavg2k8,'max day volume')
```

```python
# Plot the data
plt.plot(returns_day2k8.index, returns_day2k8)

# Add chart labels and title
plt.xlabel('Date')
plt.ylabel('Returns')
plt.title('Returns in 2008')

# Add vertical lines for max, min, and mean values
plt.axhline(y=returns_daymax2k8, color='red', linestyle='--', label='Max')
plt.axhline(y=returns_daymin2k8, color='green', linestyle='--', label='Min')
plt.axhline(y=returns_dayavg2k8, color='orange', linestyle='--', label='Mean')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
