## cancelled orders data set extraction 

```python
# Filter the cancelled orders
c = amazon[amazon['Status'] == 'Cancelled'].value_counts().reset_index()
c
```

```python
cancelled_states = amazon[amazon['Status'] == 'Cancelled']['ship-state'].value_counts()
print(cancelled_states)
```

```python
canclled_items=c.groupby('Category')['Qty'].sum().reset_index()
canclled_items
```

```python

# Combine the two DataFrames into a single DataFrame for plotting
merged_df = sum_qty.merge(canclled_items, on='Category', suffixes=('_total', '_cancelled'))
plt.style.use('dark_background')

# Create the bar plot
sns.set(style='whitegrid')
sns.barplot(x='Category', y='Qty_total', data=merged_df, color='blue', label='Total Qty')
sns.barplot(x='Category', y='Qty_cancelled', data=merged_df, color='red', label='Cancelled Qty')

# Add chart labels and legend
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.title('Comparison of Total vs Cancelled Quantities by Category')
plt.legend()

# Show the plot
plt.show()
```
